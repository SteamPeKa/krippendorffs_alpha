```python
import examples
from IPython.display import display, Markdown
```

```python
result = []

for example_name,example_info in examples.all_examples.items():
    example = example_info["example"]
    metrics_for_examples = example_info["metrics"]
    summary =example.summary(metrics=metrics_for_examples,table_processor='html',headers_level=3)
    result.append(f"## Example \"{example_name}\"\n\n{summary}")

display(Markdown("\n\n\n".join(result)))
```
